```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = [0,1,3,6,7]
Y = [1,5,2,3,7]
```

```python
plt.scatter(X,Y)
```

```python
def K_gaussian(x,y,factor)->float:
    """This function returns the evaluation of the gaussian kernel on x and y
    
    Args:
        x : float
        y : float
    Returns:
        k(x,y)
    """
    
    return np.exp(-factor*((x-y)**2))
factor = 1

def K_laplacian(x:np.ndarray,y:np.ndarray,factor:float)->float:
    """This function returns the evaluation of the laplacian kernel on x and y
    Args:
        x : float
        y : float
    Returns:
        k(x,y)
        
    """
    return np.exp(-factor*(abs(x-y)))
```

```python
kernel =lambda x,y: K_gaussian(x,y,factor)
coeffs = [[],[],[],[],[]]
for i in range(5):
    for j in range(5):
        coeffs[i].append(kernel(X[i],X[j]))  
print(coeffs)
```

```python
solution = np.linalg.solve(coeffs,Y)
solution
```

```python
def f(x):
    return solution[0]*kernel(X[0],x) + solution[1]*kernel(X[1],x) +solution[2]*kernel(X[2],x) + solution[3]*kernel(X[3],x) +solution[4]*kernel(X[4],x)
```

```python
f_X= []
X_f = []
n=100
for i in range(n):
    k = 7*i/(n-1)
    f_X.append(f(k))
    X_f.append(k)
```

```python
plt.scatter(X,Y)
plt.plot(X_f,f_X)
plt.show()
```
